```python
edges = []
num_node = 200
node_degree = 4
e_range = 0.2
for x in range(num_node):
    node_range = int(num_node*e_range)
    upper = 0
    lowwer = 0
    if (x-node_range/2<0):
        up = node_range
        low = 0
    elif (x+node_range/2>num_node-1):
        up = num_node-1
        low = int(num_node-node_range)
    else: 
        up = int(round(x+node_range/2))
        low = int(round(x-node_range/2))
        if (up>num_node-1):
            up = num_node-1
        if (low<0):
            low = 0
    e = []
    while(len(e)< node_degree):
        i = randint(low, up)
        if (i not in e and i!= x):
            e.append(i)
    for j in e:
        edges.append((x,j))
for e in edges:
        if ((e[1], e[0]) in edges):
            edges.remove((e[1], e[0]))
```
