```python
import numpy as np
import random
from random import randint
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
plt.rcParams["mathtext.fontset"] = "cm"

from collections import OrderedDict
plt.rcParams["mathtext.fontset"] = "cm"

linestyle_str = OrderedDict([
     ('solid', 'solid'),      # Same as (0, ()) or '-'
     ('dotted', 'dotted'),    # Same as (0, (1, 1)) or '.'
     ('dashed', 'dashed'),    # Same as '--'
     ('dashdot', 'dashdot')]  # Same as '-.'
)
linestyles = OrderedDict([
     ('loosely dotted',        (0, (1, 10))),
     ('dotted',                (0, (1, 1))),
     ('densely dotted',        (0, (1, 1))),

     ('loosely dashed',        (0, (5, 10))),
     ('dashed',                (0, (5, 5))),
     ('densely dashed',        (0, (5, 1))),

     ('loosely dashdotted',    (0, (3, 10, 1, 10))),
     ('dashdotted',            (0, (3, 5, 1, 5))),
     ('densely dashdotted',    (0, (3, 1, 1, 1))),

     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))]
)
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
```

```python
from IPython.display import HTML, display

def set_background(color):    
    script = (
        "var cell = this.closest('.jp-CodeCell');"
        "var editor = cell.querySelector('.jp-Editor');"
        "editor.style.background='{}';"
        "this.parentNode.removeChild(this)"
    ).format(color)

    display(HTML('<img src onerror="{}">'.format(script)))
xlabel_fontsize = 30
ylabel_fontsize = 30
xticks_fontsize = 26
yticks_fontsize = 26
```

```python
# total cost vs. # SIoT

x = [200,300,400,500,600]
# y = [800,1200,1600,2000,2400]
y_t = [1.0, 1.4, 1.8, 2.2]
T_PFL =  [1.5502800000000002, 1.38814, 1.12186, 1.03317, 0.95649]
T_TFL =  [2.21676, 2.17193, 2.15654, 2.1590699999999997, 2.16085]
T_HFL =  [2.1899300000000003, 2.04944, 1.9218, 1.84514, 1.78309]
T_SFL =  [1.7680900000000001, 1.7265300000000002, 1.66393, 1.54135, 1.44296]

fig, ax = plt.subplots(figsize=(6, 4), dpi=72)
xlabel_fontsize = 28
ylabel_fontsize = 28
xticks_fontsize = 24
yticks_fontsize = 24

# plt.axhline(y = 0.800, ls = "-", color ="lightgrey", lw=1)
plt.axhline(y = 1.200, ls = "-", color ="lightgrey", lw=1)
plt.axhline(y = 1.600, ls = "-", color ="lightgrey", lw=1)
plt.axhline(y = 2.000, ls = "-", color ="lightgrey", lw=1)


Data_1, = plt.plot(x, T_PFL,'-.^',label='PFL', color="steelblue", lw=2, ms=11) #畫線
Data_2, = plt.plot(x, T_TFL, '--*',label='TFL', color ="yellowgreen", lw=2, ms=11) #畫線
Data_3, = plt.plot(x, T_HFL, ':s',label='HFL', color="grey", lw=2, ms=11) #畫線
Data_4, = plt.plot(x, T_SFL, 'p',label='SFL', color="darkorange", linestyle = linestyles["dashdotdotted"], lw=2, ms=11) #畫線
leg = plt.legend(handles=[Data_1, Data_2, Data_3, Data_4], loc=10, bbox_to_anchor=(0.5, 1.1), prop={'size': '21', 'family': 'Times New Roman'},frameon='False', labelspacing=0, handletextpad=0.2, handlelength=1, columnspacing=0.5, ncol=5,facecolor='None')
leg.get_frame().set_linewidth(0.0)

plt.xlim(150, 650)
plt.xlabel("Number of SIoTs", fontsize=xlabel_fontsize, family="Times New Roman")
plt.xticks(fontsize=xticks_fontsize, family="Times New Roman")
# plt.xticks(np.arange(100,1100,200))
plt.yticks(np.arange(0.8, 2.4001, 0.4))
plt.yticks(fontsize=yticks_fontsize, family="Times New Roman")
plt.ylabel("Total cost(K)",fontsize=ylabel_fontsize, family="Times New Roman")

# plt.tick_params(axis='both', labelsize=22, color='green')

loc = "Figure/Total_cost_vs_nSIoT.pdf"
plt.savefig(loc,bbox_inches = 'tight')
loc = "Figure/Total_cost_vs_nSIoT.png"
plt.savefig(loc)
loc = "Figure/Total_cost_vs_nSIoT.eps"
plt.savefig(loc)

plt.show() #顯示繪製的圖形
```
